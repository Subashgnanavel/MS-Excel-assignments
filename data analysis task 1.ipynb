```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
Data = {'Name': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
'Age': [25, 30, 35, 40, 29,45,39,33],
'Department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR'],
 'Salary': [50000, 60000, 70000, 80000, 52000,80500,79888,54330],
 'JoinDate': ['2019-04-01', '2020-05-10', '2018-07-23', '2021-08-15', '2019-11-11', '2019-08-13',
'2022-08-17', '2023-12-11',]
}
```

```python
df=pd.DataFrame(Data)
df.to_csv("Employee_Data.csv",index=False)
df
```

```python
high_salary = df[(df['Department'] == 'IT')&(df['Salary']>60000)]
high_salary
```

```python
df['Experience'] = (datetime.now() - pd.to_datetime(df['JoinDate'])).dt.days // 365
df
```

```python
print(df.groupby('Department')[['Salary', 'Age']].mean())
```

```python
print(df.nlargest(3,'Salary'))
```

```python
df['Tax'] = df['Salary'].apply(lambda x: x * 0.1 if x < 60000 else x*0.2)
df
```

```python
sns.histplot(df['Age'] ,kde=True )
plt.show()
```

```python
sns.scatterplot(x='Experience',y='Salary', hue= "Department", data=df)
plt.show()
```
